```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
giraffes=cv2.imread("./giraffes.jpg",0)
giraffes_show=cv2.cvtColor(giraffes,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(giraffes_show)
```

```python
ret,thresh1 = cv2.threshold(giraffes,110,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
plt.imshow(hsv_giraffe)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
#creating a low pass filter with a 4x4 Kernel
```

```python
kernel=np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
dst=cv2.filter2D(giraffes_show,-1,kernel)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
display_img(dst)
```

```python
gaussian_blur= cv2.blur(dst,ksize=(5,5))
```

```python
plt.imshow(gaussian_blur)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
sobelx = cv2.Sobel(giraffes,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx)
```

```python
# Plotting color histogram for giraffe
```

```python
img = giraffes
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Giraffes Histograms')
plt.show()
```
